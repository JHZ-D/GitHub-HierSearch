```python
dirpath = 'Github-Links/'
links = []
for i in range(452048, 516071):
    try:
        with open(dirpath + str(i), 'r') as ipfile:
            link = ipfile.readline()
            if link != '\n':
                links.append((i, link))
    except IOError:
        print(i)
with open('Dirrepos2.txt', 'w') as opfile:
    for link in links:
        opfile.write(str(link[0]) + ' ' + link[1])
```

```python
links = [it for it in links if it != '\n']
with open('Dirrepos.txt', 'w') as opfile:
    for link in links:
        opfile.write(link)
```

```python
with open('DirGit.txt', 'w') as opfile:
    for link in links:
        if link.find('github.com') != -1:
            opfile.write(link)
```

```python
with open('DirGit.txt', 'w') as opfile:
    for link in links:
        if link.find('github.com') != -1:
            if link.endswith('.git\n'):
                opfile.write(link[:-5] + '\n')
            else:
                opfile.write(link)
```

```python
# links = [it for it in links if link.find('github.com') != -1]
links = []
with open('DirGit.txt', 'r') as ipfile:
    for link in ipfile:
        p = link.find('github.com/')
        str_exp = link[p + 11:]
        if str_exp.find('/') == -1 or len(str_exp) == str_exp.find('/') + 2:
            continue
        sp = str_exp.find('/', str_exp.index('/') + 1)
        if sp != -1:
            links.append(link[:p + 11 + sp])
        else:
            links.append(link[:-1])
with open('DirGitUnif.txt', 'w') as opfile:
    for link in links:
        opfile.write(link + '\n')
# links1 = []
# for link in links:
#     if link.endswith('.git\n'):
#         links1.append(link[:-5])
#     else:
#         p = link.find('github.com/')
#         str_exp = link[p + 11:]
#         sp = str_exp.index('/', str_exp.index('/') + 1)
```
