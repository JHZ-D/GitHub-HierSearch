```python
def unmark_element(element, stream=None):
    if stream is None:
        stream = StringIO()
    if element.text:
        stream.write(element.text)
    for sub in element:
        unmark_element(sub, stream)
    if element.tail:
        stream.write(element.tail)
    return stream.getvalue()


# patching Markdown
Markdown.output_formats["plain"] = unmark_element
__md = Markdown(output_format="plain")
__md.stripTopLevelTags = False


def unmark(text):
    return __md.convert(text)
```

```python
# func: patching Markdown

no_spaces = True #Remove spaces in filenames.
orpath = 'Extract-Readmes/'
path = 'Extract-Readmes-Paras/'
fcs = '`'   # fenced code sign
snum = 0    # sign number

fcss = '`'   # fenced code sign
snumm = 0    # sign number

def make_dir(proc_path):
    if not os.path.exists(proc_path):
        os.makedirs(proc_path)

def clean_file_name(fname):
    fname = fname.replace(":", "-")
    fname = fname.replace("/", "_")
    fname = fname.replace("\\", "_")
    fname = fname.replace("?", "-")
    fname = fname.replace("|", "-")
    fname = fname.replace("—", "-")
    if (no_spaces):
        fname = fname.replace(" ", "-")
    return fname

def print_file(subpath, fname, text):
    fname = clean_file_name(fname)
    subpath = clean_file_name(subpath)
    # make_dir(path + subpath)
    # subpath += "/"

    text_file = open(path + subpath, "w", encoding='utf-8')
    text_file.write(text)
    text_file.close()

def thematic_break(line, para):
    if para == 1 and (re.fullmatch(r' {0,3}=+ *', line) or re.fullmatch(r' {0,3}-+ *', line)):
        return 4
    if re.match(r'^ {0,3}-', line):
        if line.count('-') < 3 or re.search(r'[^ -]', line):
            return 0
        return 2
    if re.match(r'^ {0,3}_', line):
        if line.count('_') < 3 or re.search(r'[^ _]', line):
            return 0
        return 2
    if re.match(r'^ {0,3}\*', line):
        if line.count('*') < 3 or re.search(r'[^ \*]', line):
            return 0
        return 2
    return 0

def ATX(line):
    if re.match(r'^ {0,3}#{1,6}[ ]', line) or re.fullmatch(r' {0,3}#{1,6}', line):
        return 1
    return 0

def Fenced_code_bg(line):
    global fcss
    global snumm
    if re.match(r'^ {0,3}`{3,}', line):
        if re.match(r'^ {0,3}`{3,}[^`]+`', line):
            return 0
        fcss = '`'
        snumm = line.rfind('`') - line.find('`') + 1
        return 1
    if re.match(r'^ {0,3}~{3,}', line):
        fcss = '~'
        i = line.find('~')
        l = len(line)
        for j in range(i, l):
            if line[j] != '~':
                break
        snumm = j - i + (1 if line[j] == '~' else 0)
        return 1
    return 0

def table_bg(line0, line1):
    line0 = re.sub(r'\\\|', '', line0).strip()
    line1 = re.sub(r'\\\|', '', line1).strip()
    t = line0.find('|')
    if t == -1:
        return 0
    line0 = re.sub(r'(^\|)|(\|$)', '', line0)
    line1 = re.sub(r'(^\|)|(\|$)', '', line1)
    if line0.count('|') != line1.count('|'):
        return 0
    cells = line1.split('|')
    for cell in cells:
        cell = cell.strip()
        if cell == '':
            return 0
        if cell[0] == ':':
            cell = cell[1:]
        if cell == '':
            return 0
        try:
            if cell[-1] == ':':
                cell = cell[:-1]
        except:
            print(line0)
            print(line1)
            print(cell)
            quit()
        if re.search(r'[^-]', cell):
            return 0
    return 1

def chooseblock(line0, para):
    if re.match(r' {0,3}>', line0):
        return 0   # block quote
    t = thematic_break(line0, para)
    if t != 0:
        return t   # 2:thematic break, 4:Setext
    if para == 1:
        if re.match(r' {0,3}[-\+\*] +\S', line0) or re.match(r' {0,3}1[\.\)] +\S', line0):
            return 1
    else:
        if re.match(r' {0,3}[-\+\*] ', line0) or re.fullmatch(r' {0,3}[-\+\*]', line0) or re.match(
            r' {0,3}[0-9]{1,9}[\.\)] ', line0) or re.fullmatch(r' {0,3}[0-9]{1,9}[\.\)]', line0):
            return 1

    if ATX(line0):
        return 3   # ATX
    
    if re.match(r' {4,}', line0):
        return 5   # indented code block
    if Fenced_code_bg(line0):
        return 6   # fenced code block
    if not re.search(r'\S', line0):
        return 11  # blank line
    return 9       # paragraph
                   # 10 list
    
def writ(tree, j, o, l):
    if j == 0:
        tree.append([[o, l]])
        return
    writ(tree[-1], j - 1, o, l)

def clse(tree, j, l):   # 树的当前这一支j层以下全部关闭
    if len(tree[0]) == 3:
        return
    if j == 0:
        if isinstance(tree[0], list):
            if tree[0][1] > l:
                tree.clear()
                return
            tree[0].append(l)
            if len(tree) > 1:
                opntcls(tree[-1], l)
        else:
            tree.append(l)
        return
    if len(tree) > 1:
        clse(tree[-1], j - 1, l)
        if tree[-1] == []:
            tree.pop()
    return

def opntcls(tree, l):  # 把某棵树全部关闭
    if len(tree[0]) == 3:
        return
    tree[0].append(l)
    if len(tree) > 1:
        opntcls(tree[-1], l)
    return

def mdf(tree, i, o):
    if i == 0:
        tree[0][0] = o
        return tree[0][2] - tree[0][1]
    return mdf(tree[-1], i - 1, o)

def preblk(line, b):   # 开头空格数是否>=b
    l = len(line) - len(line.lstrip())
    return 0 if l < b else 1      

def fenced(line):
    line = line.strip()
    l = len(line)
    if l >= snum and line == fcs * l:
        return 1
    return 0
    
def schp(s):
    yq = re.search(r'[^ \n]', s)
    if not yq:
        return -1
    x = yq.span()[0]
    y = x
    s = s[x:]
    if s[0] == '<':
        q = re.search(r'[^\\]>', s)
        if not q:
            return -1
        p = q.span()[0] + 1
        if re.search(r'[^\\]<', s[:p]) or s[:p].find('\n') != -1:
            return -1
        if p == len(s) - 1 or not s[p + 1] in {' ', '\n', ')'}:
            return -1
        yq = re.search(r'[^ \n]', s[p + 1:]).span()
        y += p + 1 + yq[0]
        s = s[p + 1 + yq[0]:]
    else:
        q = s.find(' ')
        q1 = s.find('\n')
        if q == -1 or (q1 != -1 and q1 < q):
            q = q1
        if q == -1:
            s1 = s
        else:
            s1 = s[:q]
        l1 = len(s1)
        n = 0
        for i in range(l1):
            if s1[i] == '(' and (i == 0 or s1[i - 1] != '\\'):
                n += 1
            elif s1[i] == ')' and (i == 0 or s1[i - 1] != '\\'):
                n -= 1
            if n < 0:
                break
        if n > 0:
            return -1
        if n < 0:
            if s1[:i].find('\n') == -1:
                return x + i
            return -1
        if q == -1:
            return -1
        yq = re.search(r'[^ \n]', s[q:]).span()
        y += q + yq[0]
        s = s[q + yq[0]:]
    if len(s) == 1:
        if s[0] == ')':
            return y
        return -1
    if len(s) == 0:
        return -1
    if not s[0] in {'\'', '"', '('}:
        return -1
    if s[0] == '\'':
        q = re.search(r'[^\\]\'', s)
    elif s[0] == '"':
        q = re.search(r'[^\\]"', s)
    elif s[0] == '(':
        q = re.search(r'[^\\]\)', s)
    if not q:
        return -1
    p = q.span()[0] + 1
    yq = re.search(r'[^ \n]', s[p + 1:])
    if yq and s[p + 1 + yq.span()[0]] == ')':
        return y + p + 1 + yq.span()[0]
    return -1





def func(md_text):

    # md_text = ""
    # md_file = open(orpath + md_file_name, "r", encoding='utf-8')
    # md_text = md_file.read()
    # md_file.close()
    lines = md_text.split("\n")
    tborpara = ''
    blanklist = 0


    texti = [''] * 15
    refilen = ['Ref.md', '', '', 'ATX.md', 'Setext.md', 'IndentedCode.md', 'FencedCode.md', '', 'Table.md', 'Paragraph.md', 'List.md', 'autolink.md', 'codespan.md', 'link.md', 'picture.md']


        
    ln = 0
    l = len(lines)
    tree = [[-1, 0]]
    prefix = []
    pp = []
    while ln < l:
        line = lines[ln]
        line = line.expandtabs(4)
        prefix = pp
        lp = len(prefix)
        ll = len(line)
        i = 0
        j = 0
        pp = []
        treeflag = 0
        listn = 0
        refn = 0
        while j < ll:
            newline = line[j:]
            typ = chooseblock(newline, 1 if treeflag == 0 and i < lp and prefix[i] == 9 else 0)
            if treeflag == 0 and i < lp:
                if prefix[i] == 5 and (typ == 5 or typ == 11):
                    pp = prefix
                    texti[5] += newline + '\n'
                    treeflag = 1
                    break
                elif prefix[i] == 6:
                    if typ != 5 and fenced(newline):
                        clse(tree, i + 1, ln)
                        pp = prefix[:-1]
                        texti[6] += newline + '\n\n'
                    else:
                        pp = prefix
                        texti[6] += newline + '\n'
                    treeflag = 1
                    break
                elif typ == 11: # 空行
                    ii = i
                    while ii < lp:
                        if prefix[ii] == 10:
                            ii += 2
                        else:
                            break
                    if ii >= lp and isinstance(prefix[-1], list) and blanklist:
                        clse(tree, len(prefix), ln - 1)
                        pp = prefix[:-1]
                    else:
                        clse(tree, ii + 1, ln - 1)
                        pp = prefix[:ii]
                    if prefix[i] == 10 and listn == 0:
                        texti[10] += '\n'
                    treeflag = 1
                    break
                elif typ == 0 and prefix[i] == 0:   # 块引用与上一行对齐
                    i += 1
                    j += newline.find('>') + 1
                    if j < ll and line[j] == ' ':
                        j += 1
                    pp.append(0)
                    refn += 1
                    if refn == 1:
                        texti[0] += newline + '\n'
                    continue
                elif prefix[i] == 8 and (typ == 8 or typ == 9):
                    pp = prefix
                    texti[8] += newline + '\n'
                    treeflag = 1
                    break
                elif prefix[i] == 10 and i < lp - 1 and isinstance(prefix[i + 1], list) and preblk(newline, prefix[i + 1][1]):             # 上一行这一层是列表项，缩进以纳入列表项
                    b = prefix[i + 1][1]
                    i += 2
                    j += b
                    pp.append(10)
                    pp.append([1, b]) #!!!
                    listn += 1
                    if listn == 1:
                        texti[10] += newline + '\n'
                    continue
                elif prefix[-1] == 9 and (typ == 9 or typ == 5): # 与上一行是一个段落
                    if prefix.count(10) != 0 and pp.count(10) == 0:
                        texti[10] += newline + '\n'
                    if prefix.count(0) != 0 and pp.count(0) == 0:
                        texti[0] += newline + '\n'
                    x = table_bg(tborpara, newline)  # 表格
                    if x and prefix[i] == 9:
                        clse(tree, i + 1, ln - 2)
                        writ(tree, i, 8, ln - 1)
                        texti[9] = texti[9][:-1]
                        ps = texti[9].rfind('\n')
                        texti[9] = texti[9][:ps + 1]
                        texti[8] += '\n' + tborpara + '\n' + newline + '\n'
                        pp = prefix[:-1]
                        pp.append(8)
                    else:
                        pp = prefix
                        texti[9] += newline + '\n'
                        tborpara = newline
                    treeflag = 1
                    
                    break

                if prefix[-1] == 5:
                    texti[5] += '---\n'
                elif prefix[-1] == 6:
                    texti[6] += fcs * snum + '\n'
                if prefix[i] == 10 and typ == 1:
                    clse(tree, i + 2, ln - 1)
                else:
                    clse(tree, i + 1, ln - 1)
            treeflag = 1
            if typ == 1:
                lline = newline.lstrip()
                b = len(newline) - len(lline)
                b1 = lline.find(' ')
                if b1 == -1:
                    b += len(lline) + 1
                else:
                    lline = lline[b1:]
                    if lline.strip() == '':
                        b2 = 1
                    else:
                        b2 = len(lline) - len(lline.lstrip())
                    if b2 > 4:
                        b += b1 + 1
                    else:
                        b += b2 + b1
                listn += 1
                if i < lp and prefix[i] == 10:
                    writ(tree, i + 1, 1, ln)
                else:
                    writ(tree, i, 10, ln)
                    writ(tree, i + 1, 1, ln)
                    if listn == 1:
                        texti[10] += '\n'
                pp.append(10)
                pp.append([1, b])
                if listn == 1:
                    texti[10] += newline + '\n'
                j += b
                if re.fullmatch(r' *', line[j:]):
                    blanklist = 1
                else:
                    blanklist = 0
                    
                i += 1

            elif typ == 0:
                writ(tree, i, 0, ln)
                pp.append(0)
                j += newline.find('>') + 1
                if j < ll and line[j] == ' ':
                    j += 1
                refn += 1
                if refn == 1:
                    texti[0] += '\n' + newline + '\n'

            elif typ == 2 or typ == 3:    # thematic break / ATX
                writ(tree, i, typ, ln)
                clse(tree, i + 1, ln)
                pp.append(typ)
                if typ == 3:
                    texti[3] += newline + '\n'
                break
            elif typ == 5:          # indented code block
                writ(tree, i, 5, ln)
                pp.append(typ)
                texti[5] += '\n' + newline + '\n'
                break
            elif typ == 6:   # fenced code block
                writ(tree, i, 6, ln)
                pp.append(typ)
                texti[6] += '\n' + newline + '\n'
                fcs = fcss
                snum = snumm
                break 

            elif typ == 8: # table
                writ(tree, i, 8, ln)
                pp.append(typ)
                texti[8] += '\n' + newline + '\n'
                break

            elif typ == 9:
                writ(tree, i, 9, ln)
                pp.append(typ)
                texti[9] += '\n' + newline + '\n'
                tborpara = newline    # table or paragraph?
                break
            
            elif typ == 4:
                mp = mdf(tree, i + 1, 4)
                pp.append(4)
                para = texti[9][:-1]
                pt = ''
                for mi in range(-mp - 1, 0):
                    ps = para.rfind('\n')
                    pt = para[ps + 1:] + '\n' + pt
                    para = para[:ps]
                texti[4] += pt
                texti[9] = para + '\n'
                texti[4] += newline + '\n\n'
                break
            else:
                break
            i += 1
            
        if treeflag == 0 and i < lp:
            if not (prefix[i] == 5 or prefix[i] == 6):
                ii = i
                while ii < lp:
                    if prefix[ii] == 10:
                        ii += 2
                    else:
                        break
                if ii >= lp and isinstance(prefix[-1], list) and blanklist:
                    clse(tree, lp, ln - 1)
                    pp = prefix[:-1]
                else:
                    clse(tree, ii + 1, ln - 1)
                    pp = prefix[:ii]
                if prefix[i] == 10:
                    texti[10] += '\n'
                if prefix[-1] == 6:
                    texti[6] += fcs * snum + '\n'
                elif prefix[-1] == 5:
                    texti[5] += '---\n'
            else:
                pp = prefix
                texti[prefix[-1]] += '\n'
        ln += 1
    clse(tree, 0, l - 1)
    # f = open('result.md', 'w')
    # print(tree, file=f)
    # f.close()

    parag = []
    sign = 0
    paralist = re.split(r'\n *\n', texti[9])
    paratext = ''
    for para in paralist:
        pc = re.match(r'\[.*?\]\(.*?\)', para)
        if pc:
            pp = pc.span()[1]
            if pp == len(para):
                continue
        paratext += para + '\n\n'
    return paratext
    # subpath = md_file_name
    # print_file(subpath, 'para.md', paratext)




# if not os.path.exists(path):
#     make_dir(path)
# if len(sys.argv) > 1:
#     if sys.argv[1] != "":
#         md_file_name = sys.argv[1]
#         if not os.path.exists(orpath + md_file_name):
#             print("File missing:", md_file_name)
#             quit()
#     func(md_file_name)
# else:
#     for file in os.listdir(orpath):
#         md_file_name = file
#         if os.path.exists(path + md_file_name):
#             continue
#         print(md_file_name)
#         func(md_file_name)
```
